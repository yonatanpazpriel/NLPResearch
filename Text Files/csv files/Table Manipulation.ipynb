```python
import pandas as pd
import re
```

```python
# Load CSV file into a DataFrame
texts = pd.read_csv('texts.csv')
texts
```

```python
def extract_name(filename):
    pattern = r'\d{1,2}-\d{1,2}-\d{2} (\w+)_embedded\.txt'
    match = re.search(pattern, filename)
    if match:
        # Convert the name to lowercase
        return match.group(1).lower()
    return None  # Return None if no match is found

# Apply the function to extract names and create a new column 'name'
texts['name'] = texts['file_name'].apply(extract_name)
texts
```

```python
def extract_full_name(row):
    last_name = row['name']
    text = row['preprocessed_text']
    
    if last_name is None:  # Check if last name is None
        return None  # Or handle it differently if needed

    # Regex to match the full name (assuming the first name precedes the last name)
    pattern = r'(\b\w+\b) ' + re.escape(last_name)  # Matches first name followed by last name
    match = re.search(pattern, text)
    
    return match.group(1) if match else None  # Return first name or None if not found


# Apply the function to create a new column 'first_name'
texts['first_name'] = texts.apply(extract_full_name, axis=1)
texts['author'] = texts['first_name'] + ' ' + texts['name']
texts
```
