```python
# sampling script
# Calculate the total submissions per folder_name
folder_totals = all_files.groupby('folder_name')['submissions'].sum()

# Calculate the proportion of each folder
folder_proportions = folder_totals / folder_totals.sum()

# Assign proportional weights to each row
all_files['weight'] = all_files['folder_name'].map(folder_proportions) * all_files['submissions']

# Normalize the weights so they sum to 1
all_files['weight'] = all_files['weight'] / all_files['weight'].sum()

# Sample 300 rows based on these weights
sampled_all = all_files.sample(n=300, weights='weight', random_state=18, replace=True)

# Reset the index for the sampled DataFrame
sampled_all.reset_index(drop=True, inplace=True)

# Display the sampled DataFrame
sampled_all
```

```python
import os

# Initialize a list to store the file information
file_data = []

# Set the path to the directory
root_dir = './sampled_pdfs'

# Walk through the directory
for dirpath, dirnames, filenames in os.walk(root_dir):
    for filename in filenames:
        if filename.lower().endswith('.pdf'):
            # Get the folder name (immediate parent folder)
            folder_name = os.path.basename(dirpath)
            # Append the data to the list
            file_data.append({'file_name': filename, 'folder_name': folder_name})

# Create a DataFrame from the list
temp = pd.DataFrame(file_data)
```
